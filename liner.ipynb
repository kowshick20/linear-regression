```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pylab as pl
```

```python
data=pd.read_csv("FuelConsumptionCo2.csv")
data.head()
```

```python
data.describe()
```

```python
cdf = data[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

```python
data['CO2EMISSIONS']=data['CO2EMISSIONS'].astype(float)
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB,cdf.CO2EMISSIONS,color='blue')
plt.xlabel('fuel consumption')
plt.ylabel('co2 emission')
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS,cdf.CO2EMISSIONS,color='blue')
plt.xlabel('cylindes')
plt.ylabel('co2 emission')
plt.show()
```

```python
data.info()
```

```python
msk = np.random.rand(len(data)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
train_x=np.asanyarray(train[['ENGINESIZE']])
train_y=np.asanyarray(train[['CO2EMISSIONS']])
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(train_x,train_y)
print('coefficient:',model.coef_)
print('intercept',model.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, model.coef_[0][0]*train_x + model.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import mean_absolute_error as mae
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = model.predict(test_x)

print("Mean squared error: %.2f" % rmse(test_x,test_y))
print("mean absolut error: %.2f" % mae(test_x,test_y) )
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```
